**Campaign Performance :**

**âž¢ Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.**

```python
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
# relationship between Acquisition_cost vs ROI using Scotter plot

plt.figure(figsize=(10,8),facecolor='pink')
plt.scatter(data['Acquisition_Cost'],data['ROI'],color='purple')
plt.title('Acquisition_cost vs ROI',fontsize=25,color='red')
plt.xlabel('Acquisition_Cost',color='red',fontsize=25)
plt.ylabel('ROI',color='red',fontsize=25)
plt.grid(True)
plt.show()
```

**Explanation:**

* X - axis represents Acquitision_cost.   
* Y - axis represents ROI(Return on Ivestment).
* Acqutision_cost is high the ROI is low so there is a negative correlation between these two columns.

**Time-Based Analysis:**

**âž¢ Plot the distribution of Duration using a histogram.**

In the dataframe duartions are in the categorical so we convert into numerical by the label encoder

```python
#convert duration into numeric value

from sklearn.preprocessing import LabelEncoder

data_encoded=data.copy()

Label_encoder = LabelEncoder()

data_encoded['Duration']=Label_encoder.fit_transform(data_encoded['Duration'])

data_encoded['Duration']
```

```python
# distribution of Duration using a histogram
plt.figure(figsize=(10,6),facecolor='pink')
sns.histplot(data_encoded['Duration'],bins=5,color='pink',edgecolor='black',kde=True)
plt.title('Distribution of Duration',color='red',fontsize=20)
plt.xlabel('Duration',color='red',fontsize=20)
plt.ylabel('Count',color='red',fontsize=20)
plt.show()
```

Explanation:

* X-axis representsDuration
* Y-axis represents Count
* It shows the histogram of distribution of  the duration.
* we convert the duration categorical to numerical with help of label encoder
* The encoder duration splited like short time,median time and long time
* Over all this histogram we understand the count of the skew is high in the median time of duration.
